Buatlah suatu program sederhana untuk menyelesaikan :

<ol>
<li>Mencari Kontur Interval (CI) dengan memberikan masukan oleh user berupa penyebut skala </li>
<li>Mencari mencari nilai ketinggian(h) dan kecepatan(v) setelah bola dilemparkan dengan h0 = 8 m dan v0 = 2 m/s dengan variasi t yakni 0.5 s, lebih dari 0.5 s, dan lebih dari 2s. Besar g =9.8 </li>
</ol>

$$h = h_0 + V_0t-\frac12 gt^2$$

$$v = v_0 - gt$$

Script code :

```python
# 1.Mencari Kontur Interval (CI)

print ("\033[1m" + "\n\nMencari Kontur Interval (CI)" + "\033[0m")
print (" ")
skala = input("Masukkan skala (1:x), x = ")
print (" ")
skala = str(skala)
print("Memilih Skala 1:" + skala)
print(" ")

skala = float(skala)
kontur = 0.0005 * skala

print ("Hasil Kontur Interval = " + str(kontur) + " meter")
```

```python
# 2. mencari nilai ketinggian dan kecepatan
print ("\033[1m" + "\n\nProgram Mencari Nilai Ketinggian (h) dan Kecepatan (v)" + "\033[0m")
print(" ")

#ketika t = 0.5 s
t = 0.5
h = (8+2*t)-(0.5*9.8*t**2)
v = 2-(9.8*t)
print("Ketinggian benda ketika t=" + str(t) + " -> " + " %.2f " %h)
print("Kecepatan benda ketika t=" + str(t) + "  -> " + " %.2f " %v)
print(" ")

#ketika t = 1.5 s
t = 1.5
h = (8+2*t)-(0.5*9.8*t**2)
v = 2-(9.8*t)
print("Ketinggian benda ketika t=" + str(t) + " -> " + " %.2f " %h)
print("Kecepatan benda ketika t=" + str(t) + "  -> " + " %.2f " %v)
print(" ")

#ketika t = 3 s
t = 3
h = (8+2*t)-(0.5*9.8*t**2)
v = 2-(9.8*t)
print("Ketinggian benda ketika t=" + str(t) + "   -> " + " %.2f " %h)
print("Kecepatan benda ketika t=" + str(t) + "    -> " + " %.2f " %v)
```

Script code:

```python
import pandas as pd
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from matplotlib.pyplot import scatter
```

```python
Data = pd.read_excel('C:\\document\\praktikum python\\Data Gempa BMKG.xlsx')
```

```python
from matplotlib.image import imread
```

```python
plt.figure(figsize=(10,9))
plt.axes(projection=ccrs.PlateCarree())
plt.imshow(imread('C:\\document\\praktikum python\\Gambar.TIF'), extent=[-180, 180, -90, 90])
plt.xlim([min(Data['Longitude']),max(Data['Longitude'])])
plt.ylim([min(Data['Latitude']),max(Data['Latitude'])])
plt.scatter(Data['Longitude'],Data['Latitude'],c='red')
plt.title('Sebaran Gempa di Indonesia')
plt.show()
```
